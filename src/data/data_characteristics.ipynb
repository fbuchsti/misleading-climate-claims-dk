# DATA CHARACTERISTICS

```python
"""
Dataset Examination Script for Data Collection Section
This script examines the dataset and outputs all statistics needed for section 3.1
"""
```

```python
import pandas as pd
import numpy as np
```

```python
# load data
df = pd.read_csv("/Users/FrederikkeB/Documents/GitHub/misleading-climate-claims-dk/data/raw/climate_misinfo_2023-2025.csv")
df.head(2)
```

```python
# corpus size
print(f"Total articles in dataset: {len(df):,}")
```

```python
# article length statistics 
if 'wordcountbodytext' in df.columns:
    word_counts = df['wordcountbodytext']
else:
    print("ERROR: Could not find 'wordcountbodytext' column.")
    print("Available columns:", df.columns.tolist())
    word_counts = None
 
if word_counts is not None:
    print(f"Mean:   {word_counts.mean():,.1f} words")
    print(f"SD:     {word_counts.std():,.1f} words")
    print(f"Median: {word_counts.median():,.1f} words")
    print(f"Min:    {word_counts.min():,} words")
    print(f"Max:    {word_counts.max():,} words")
```

```python
# outlet distribution
if 'Media' in df.columns:
    outlet_counts = df['Media'].value_counts()
    total_outlets = len(outlet_counts)
    
    print(f"Total number of distinct outlets: {total_outlets}")
    print(f"\nTop 10 outlets by article count:")
    print("-"*70)
    
    for i, (outlet, count) in enumerate(outlet_counts.head(10).items(), 1):
        percentage = (count / len(df)) * 100
        print(f"{i:2d}. {outlet:40s} {count:6,} ({percentage:5.2f}%)")
```

```python
if 'Media' in df.columns and word_counts is not None:
    outlet_counts = df['Media'].value_counts()
    total_outlets = len(outlet_counts)
    
    print(f"Total number of distinct outlets: {total_outlets}")
    
    # Overall statistics (for the first row of the table)
    print("\n" + "="*70)
    print("TABLE 3.1: CHARACTERISTICS OF ANALYTICAL SAMPLE")
    print("="*70)
    print(f"{'Outlet':<40} {'N':>8} {'%':>7} {'Mean':>7} {'SD':>7} {'Median':>7} {'Min':>7} {'Max':>7}")
    print("-"*120)
    
    # Overall corpus row (bolded in your thesis)
    overall_mean = word_counts.mean()
    overall_sd = word_counts.std()
    overall_median = word_counts.median()
    overall_min = word_counts.min()
    overall_max = word_counts.max()
    
    print(f"{'OVERALL CORPUS':<40} {len(df):>8,} {100.0:>6.1f}% {overall_mean:>7.0f} {overall_sd:>7.0f} {overall_median:>7.0f} {overall_min:>7,} {overall_max:>7,}")
    print("-"*120)
    
    # Top 10 outlets with statistics
    top_outlets = outlet_counts.head(10)
    for outlet, count in top_outlets.items():
        percentage = (count / len(df)) * 100
        outlet_data = df[df['Media'] == outlet]['wordcountbodytext']
        
        outlet_mean = outlet_data.mean()
        outlet_sd = outlet_data.std()
        outlet_median = outlet_data.median()
        outlet_min = outlet_data.min()
        outlet_max = outlet_data.max()
        
        # Truncate outlet name if too long
        outlet_display = outlet[:38] if len(outlet) > 38 else outlet
        
        print(f"{outlet_display:<40} {count:>8,} {percentage:>6.1f}% {outlet_mean:>7.0f} {outlet_sd:>7.0f} {outlet_median:>7.0f} {outlet_min:>7,} {outlet_max:>7,}")
    
    # Aggregate "Other outlets" row
    if total_outlets > 10:
        remaining_count = len(df) - outlet_counts.head(10).sum()
        remaining_pct = (remaining_count / len(df)) * 100
        
        # Get data for all other outlets
        top_outlet_names = top_outlets.index.tolist()
        other_data = df[~df['Media'].isin(top_outlet_names)]['wordcountbodytext']
        
        other_mean = other_data.mean()
        other_sd = other_data.std()
        other_median = other_data.median()
        other_min = other_data.min()
        other_max = other_data.max()
        
        other_label = f"Other outlets (n={total_outlets - 10})"
        print(f"{other_label:<40} {remaining_count:>8,} {remaining_pct:>6.1f}% {other_mean:>7.0f} {other_sd:>7.0f} {other_median:>7.0f} {other_min:>7,} {other_max:>7,}")
    
    
    # Additional breakdowns by media type and region
    if 'Medietype' in df.columns:
        print("\n" + "="*70)
        print("DISTRIBUTION BY MEDIA TYPE")
        print("="*70)
        type_counts = df['Medietype'].value_counts()
        for media_type, count in type_counts.items():
            percentage = (count / len(df)) * 100
            print(f"{media_type:<40} {count:>15,} {percentage:>9.1f}%")
    
    if 'MediaRegion' in df.columns:
        print("\n" + "="*70)
        print("DISTRIBUTION BY MEDIA REGION")
        print("="*70)
        region_counts = df['media region'].value_counts()
        for region, count in region_counts.items():
            percentage = (count / len(df)) * 100
            print(f"{region:<40} {count:>15,} {percentage:>9.1f}%")
    
else:
    print("ERROR: Could not find 'Media' column or word count data.")
    print("Available columns:", df.columns.tolist())
```

```python
# temporal distribution
if 'Year' in df.columns:
    year_counts = df['Year'].value_counts().sort_index()
    
    print(f"Articles by year:")
    for year, count in year_counts.items():
        percentage = (count / len(df)) * 100
        print(f"  {year}: {count:6,} ({percentage:5.2f}%)")
else:
    print("ERROR: Could not find 'Year' column.")
    print("Available columns:", df.columns.tolist())
```
